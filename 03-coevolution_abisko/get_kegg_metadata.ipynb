```python
def get_child(input_df):
    output_list = []
#     for child_1 in input_df.itertuples():
    input_df = input_df.children.values.tolist()
#     parent_metabolism = input_df['name'][0]
    for parent_1 in input_df:
        parent = parent_1['name'].split()
        parent_metabolism_id = parent[0]
        parent_metabolism = ' '.join(parent[1:])
        
        for metab_1 in parent_1['children']:
            metab = metab_1['name'].split()
            child_metabolism_id = metab[0]
            child_metabolism = ' '.join(metab[1:])
            for path_row in metab_1['children']:
                path_id = path_row['name'].split()
                path_metabolism_id = path_id[0]
                path_metabolism = ' '.join(path_id[1:])
                try:
                    for kegg_id in path_row['children']:
                        try:
                            kegg, kegg_info = kegg_id['name'].split(';')
                            kegg = kegg.split()
                            output_list.append([kegg[0], ' '.join(kegg[1:]), kegg_info, path_metabolism, path_metabolism_id, child_metabolism, child_metabolism_id, parent_metabolism, parent_metabolism_id])
                        except ValueError:
                            kegg = kegg_id['name']
                            output_list.append([kegg, ' ', ' ', path_metabolism, path_metabolism_id, child_metabolism, child_metabolism_id, parent_metabolism, parent_metabolism_id])
                except KeyError:
                    continue
                    
    output_df = pd.DataFrame(output_list, columns = ['KO', 'ID', 'INFO', 'PATH', 'PATH_ID', 'MAJOR_PATH', 'MAJOR_PATH_ID', 'META_PATH', 'META_PATH_ID'])
    
    return output_df
                    
                    
```

```python
keggs = get_child(dat)
```
