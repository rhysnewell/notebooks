```r
source("setup_network.R")
library(dplyr)
library(data.table)
library(ggplot2)
library(ggpubr)
library(grid)
library(gridExtra)
library(ape)
library(propr)

###1. Read in matrices
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
###2. Creating gml files of network (to be visulized in Gephi or Cytoscape)
pattern<-co_occurrence_network(scale(trimmed_abundances_ten), 0.6,0.01)  ## cutoffs for correlation coefficient and P-value

write.graph(pattern$graph1,'Pos0.6-NW.gml',format='gml')    #network file for positive association
```

```r
###3. Calculating network topological properties
g<-pattern$graph1   ###positive network
#g<-pattern$graph1   ###negative network

c <- cluster_walktrap(g)
# Global toplogical features
modularity(c)
md <- modularity(g, membership(c), weights = NULL)
cc <- transitivity(g, vids = NULL,
                   weights = NULL)
spl <- average.path.length(g, directed=FALSE, unconnected=TRUE)
gd  <- graph.density(g, loops=FALSE)
nd  <- diameter(g, directed = FALSE, unconnected = TRUE, weights = NA)

node.degree <- degree(g, v = V(g), mode="all")
ad  <- mean(node.degree)

e <- ecount(g)
v <- vcount(g)
global.topology <- data.frame(e,v,cc,spl,md,gd,nd,ad)
write.csv(global.topology, file="Pos0.6-NW-global.topology.csv")

# Node toplogical features
betweenness.centrality <- betweenness(g, v=V(g), 
                                      directed = FALSE, weights = NA,
                                      nobigint = TRUE, normalized = FALSE)
closeness.centrality <- closeness(g, vids = V(g),
                                  weights = NA, normalized = FALSE)
node.transitivity <- transitivity(g, type = c("local"), vids = NULL,
                                  weights = NA)

node.topology <- data.frame(node.degree, betweenness.centrality, closeness.centrality, node.transitivity)
write.csv(node.topology, file="Pos0.6-NW-node.topology.csv")

# Ploting node degreee distribution in a log-log plot
degree.df <- data.frame(table(degree=factor(node.degree, levels=seq_len(max(node.degree)))))
degree.df$degree <- as.numeric(as.character(degree.df$degree))

#4. Creating an abundance table for OTUs present in the positive and negative network
my.list1 <- row.names(pattern$matrix.cor1)
###my.list2 <- row.names(pattern$matrix.cor2)

logical1 <- row.names(t(trimmed_abundances_ten))  %in% my.list1
###logical2 <- row.names(Abu)  %in% my.list2

tab.subset1 <- subset(t(trimmed_abundances_ten),logical1)
###tab.subset2 <- subset(Abu,logical2)
write.table(tab.subset1,'Pos0.6-NW.txt',sep="\t")
```

```r

library(SpiecEasi)
library(igraph)
library(multiplex)
#require(SpiecEasi)
#Upload FlashWeave Output into Multiplex 
#Makes a srt: send, receive, and ties
#It's a data frame with at least 3 columns for the sender, receiver, and the ties
GML_Output <- read.gml("~/Desktop/Programs/Julia/CLab3M_10_12/Combined_Lab3M_SetCore_network_output_10_12.gml", 
                       as = 'srt', directed = FALSE, coords = FALSE)

#Makes it a dataframe 
GML_Output <- as.data.frame(GML_Output)
#Makes it a graph object
GML_Output <- graph_from_data_frame(GML_Output, directed = FALSE, vertices = NULL)

#Upload OTU Table 
otu_table <- read.csv("~/Desktop/Programs/R/Core Microbiome/16S/Lab/Combined_3M_10_12/OTU_Table_FlashWeave_CLab3M_10_12.csv")
otu_table <- data.frame(otu_table, row.names = "X")

#Format data table
OTU_table <- t(otu_table)

require(igraph)
#Transform GML from a graph obeject to an adjacency matrix
GML_AdjMat <- as_adjacency_matrix(GML_Output, type = c("both", "upper", "lower"),
                                  attr = NULL, edges = FALSE, names = TRUE,
                                  sparse = 
                                    igraph_opt("sparsematrices"))

#Creating a weighted graph from an undirected adjacency matrix.
#A weighted graph is created and the name of the edge attribute will be weight
ig.n <- graph.adjacency(GML_AdjMat, mode='undirected', weighted = TRUE)

### Places the vertices randomly on a square. It has no parameters
am.coord <- layout.random(ig.n)

#Form row means for the centered log ratio transformed dataset
#Set size of vertex proportional to clr-mean
library(igraph)
vsize <- rowMeans(clr(OTU_table, 1))+6
## plot
plot(ig.n, layout=am.coord, vertex.size=vsize, vertex.label=NA, main="MB")

##### let's remove the nodes that are outside of the main component
comp_ign <- components(ig.n)
isolated <- names(comp_ign$membership[which(comp_ign$membership %in% which(comp_ign$csize < 5))])

###Function for if the lengths in isolated are greater than zero then they will be deleted from ig.n
if(length(isolated) > 0){
  ig.n <- delete_vertices(ig.n, isolated)
  data_pruned <- OTU_table[,-which(colnames(OTU_table) %in% isolated)]
  vsize <- rowMeans(clr(data_pruned, 1))+6
}else{
  vsize <- rowMeans(clr(OTU_table, 1))+6
}

### Places the vertices randomly on a square. It has no parameters
am.coord <- layout.random(ig.n)
##Plot
plot(ig.n, layout=am.coord, vertex.size=vsize, vertex.label=NA, main="MB")

##### let's see what we can say about this network
require(rnetcarto)
#Making an adjacency matrix
netc <- netcarto(as_adjacency_matrix(ig.n, attr = 'weight', sparse = FALSE))
#Places vertices on the plane using the force-directed layout algorithm by Fruchterman and Reingold
am.coord <- layout.fruchterman.reingold(ig.n)
vsize <- rowMeans(clr(OTU_table, 1))+6

V(ig.n)$module <- netc[[1]]$module[match(V(ig.n)$name, as.character(netc[[1]]$name))]

edge.weights <- function(community, network, weight.within = 100, weight.between = 1) {
  bridges <- crossing(communities = community, graph = network)
  weights <- ifelse(test = bridges, yes = weight.between, no = weight.within)
  return(weights) 
}

E(ig.n)$weight_abs <- abs(E(ig.n)$weight)
c_louv <- cluster_louvain(ig.n, weights = E(ig.n)$weight_abs)
E(ig.n)$strength <- E(ig.n)$weight
E(ig.n)$weight <- edge.weights(c_louv, ig.n)
am.coord <- layout_with_fr(ig.n)
vsize <- rowMeans(clr(OTU_table, 1))+6
#Second root transformation
vsize <- sqrt(degree(ig.n))+1

V(ig.n)$module <- as.vector(membership(c_louv))
require(RColorBrewer)
#install.packages("RColorBrewer")

custom_palette <- c('#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', 
                    '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#e6beff', 
                    '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1', 
                    '#000075', '#808080', '#ffffff', '#000000', '#8C510A', "#BF812D",
                    '#DFC27D','#F6E8C3','#C7EAE5','#80CDC1','#35978F','#01665E',
                    '#762A83','#9970AB','#1b9e77', '#d95f02','#7570b3','#8dd3c7',
                    '#ffffb3','#bebada','#e41a1c','#377eb8','#4daf4a','#7fc97f','#beaed4',
                    '#fdc086','#d8b365','#7fbf7b','#999999','#de2d26','#c51b8a','#dd1c77',
                    '#d95f0e','#2b8cbe','#a6bddb','#e6550d','#f7fcb9','#addd8e','#636363',
                    '#bdbdbd',"#feb24c")

V(ig.n)$color <- custom_palette[V(ig.n)$module]


###To add labels vertex.label=newasv$Class
## plot
###Input Network Title
#plot(ig.n, layout=am.coord, vertex.shape="pie", vertex.pie=values, vertex.size=vsize, vertex.label=NA, main="Combined Field Tetrapod Network")
plot(ig.n, layout=am.coord, vertex.size=vsize, vertex.label=c_louv$membership, main="Combined Lab3M Tetrapod Network")
```
