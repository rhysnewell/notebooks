```r
g4_all_pos <- nrow(genome4_all_core)
g4_all_neg <- 6585299 - g4_all_pos
# TP Values
g4_tp_sml <- nrow(inner_join(genome4_all_core, genome4_lorikeet_small[snv==TRUE], by=c("POS", "REF", "ALT")))
g4_tp_med <- nrow(inner_join(genome4_all_core, genome4_lorikeet_med[snv==TRUE], by=c("POS", "REF", "ALT")))
g4_tp_lrg <- nrow(inner_join(genome4_all_core, genome4_lorikeet_lrg[snv==TRUE], by=c("POS", "REF", "ALT")))

# FP values
g4_fp_sml <- nrow(genome4_lorikeet_small[snv==TRUE]) - g4_tp_sml
g4_fp_med <- nrow(genome4_lorikeet_med[snv==TRUE]) - g4_tp_med
g4_fp_lrg <- nrow(genome4_lorikeet_lrg[snv==TRUE]) - g4_tp_lrg

# TN values
g4_tn_sml <- g4_all_neg - g4_fp_sml
g4_tn_med <- g4_all_neg - g4_fp_med
g4_tn_lrg <- g4_all_neg - g4_fp_lrg

# TPR values
g4_tpr_sml <- g4_tp_sml / nrow(genome4_all_core)
g4_tpr_med <- g4_tp_med / nrow(genome4_all_core)
g4_tpr_lrg <- g4_tp_lrg / nrow(genome4_all_core)

# FPR values
g4_fpr_sml <- g4_fp_sml / (g4_fp_sml + g4_tn_sml)
g4_fpr_med <- g4_fp_med / (g4_fp_med + g4_tn_med)
g4_fpr_lrg <- g4_fp_lrg / (g4_fp_lrg + g4_tn_lrg)
```

```r
g10_all_pos <- nrow(genome10_all_core)
g10_all_neg <- 3574722 - g10_all_pos
# TP Values
g10_tp_sml <- nrow(inner_join(genome10_all_core, genome10_lorikeet_small[snv==TRUE], by=c("POS", "REF", "ALT")))
g10_tp_med <- nrow(inner_join(genome10_all_core, genome10_lorikeet_med[snv==TRUE], by=c("POS", "REF", "ALT")))
g10_tp_lrg <- nrow(inner_join(genome10_all_core, genome10_lorikeet_lrg[snv==TRUE], by=c("POS", "REF", "ALT")))

# FP values
g10_fp_sml <- nrow(genome10_lorikeet_small[snv==TRUE]) - g10_tp_sml
g10_fp_med <- nrow(genome10_lorikeet_med[snv==TRUE]) - g10_tp_med
g10_fp_lrg <- nrow(genome10_lorikeet_lrg[snv==TRUE]) - g10_tp_lrg

# TN values
g10_tn_sml <- g10_all_neg - g10_fp_sml
g10_tn_med <- g10_all_neg - g10_fp_med
g10_tn_lrg <- g10_all_neg - g10_fp_lrg

# TPR values
g10_tpr_sml <- g10_tp_sml / nrow(genome10_all_core)
g10_tpr_med <- g10_tp_med / nrow(genome10_all_core)
g10_tpr_lrg <- g10_tp_lrg / nrow(genome10_all_core)

# FPR values
g10_fpr_sml <- g10_fp_sml / (g10_fp_sml + g10_tn_sml)
g10_fpr_med <- g10_fp_med / (g10_fp_med + g10_tn_med)
g10_fpr_lrg <- g10_fp_lrg / (g10_fp_lrg + g10_tn_lrg)
```

```r
g4_df <- data.frame(tpr=c(g4_tpr_sml, g4_tpr_med, g4_tpr_lrg), fpr=c(g4_fpr_sml, g4_fpr_med, g4_fpr_lrg), sample=c("small", "med", "large"))
g4_df

ggplot(data=g4_df) + 
    geom_point(aes(x=fpr, y=tpr, colour=sample, size=3)) +
    ggtitle("Recovered Variants on Genome4") + 
  theme(
      title = element_text(size=14),
      axis.title = element_text(size=14),
    axis.text = element_text(size=14),
    legend.text = element_text(size=14),
    legend.title = element_text(size=14),
    ) 
```
