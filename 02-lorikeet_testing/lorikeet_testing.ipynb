## Get TP core SNPs using parsnp

Run from within Synth_XXX folder:

```
cat ../synth_15_paths.tsv | parallel --colsep "\t" -j 5 "mkdir -p parsnp/{1} && cp binning/data/das_tool_bins/das_tool_DASTool_bins/{1}.fa parsnp/{1}/ && cp {2} parsnp/{1}/"
```

Generate parsnp results:

```
ls parsnp/ | parallel -j5 parsnp -c -d parsnp/{}/ -r parsnp/{}/{}.fa -o parsnp/{}/snps/ -p 20
```

convert `.ggr` to `.vcf`:

```
ls parsnp/*/snps/*.ggr | parallel harvesttools -i {} -V {.}.vcf
```

### Reference only version

```
cat ../synth_15_paths.tsv | parallel --colsep "\t" -j 5 "mkdir -p parsnp_ref_only/{1} && cp {2} parsnp_ref_only/{1}/"
ls parsnp_ref_only/ | parallel -j5 parsnp -c -d parsnp_ref_only/{}/ -r ! -o parsnp_ref_only/{}/snps/ -p 20
ls parsnp_ref_only/*/snps/*.ggr | parallel harvesttools -i {} -V {.}.vcf
```
